```python
import wandb
from configs.defaults import Globs
from helpers.wandb_common import get_wandb_df
```

```python
df = get_wandb_df(project_name=Globs.project_name,
                  wandb_user=Globs.entity_name)
```

```python
filter_tags = ['tuning', 'v3']
for filter_tag in filter_tags:
    df = df[df.tags.apply(lambda x: filter_tag in x)]
```

```python
df.head(3)
```

```python
datasets, models = list(df['DATASET'].unique()), list(df['MODEL'].unique())
```

```python
# Results for the top 3 hyperparams
df.groupby(['DATASET', 'MODEL'])['final_NRMSE', 'final_ND'].apply(lambda x: x.nsmallest(3, ['final_ND']))\
    .groupby(['DATASET', 'MODEL'])['final_NRMSE', 'final_ND'].agg(['std'])
```

```python
# Results for the top hyperparams
df.groupby(['DATASET', 'MODEL'])['final_NRMSE', 'final_ND'].min()
```
